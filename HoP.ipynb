```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "purchasedata.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(file_to_load)
df.head()
```

```python
# calculate the number of unique plaYers
Player_Count=df.loc[:,['Gender','SN','Age']]
Player_Count=Player_Count.drop_duplicates()

#disp tot num players
TotalPlayers=Player_Count.count()[0]
pd.DataFrame({'TotalPlayers':[TotalPlayers]})
```

```python
#chained 
pd.DataFrame({"total Players":
             [df.loc[:,["Gender","SN","Age"]]
             .drop_duplicates()
             .count()[0]]})
```

```python
#min requirement
df.loc[:,"SN"].drop_duplicates().count()
```

```python
#making the variable 
total_players=df.loc[:,"SN"].drop_duplicates().count()
```

```python
#	Purchase ID	SN	Age	Gender	Item ID	Item Name	v
uniqueItems=len(df['Item Name'].unique())
avgPrice=round(df['Price'].mean(),2)
numPurchs=df['Purchase ID'].count()
sumPurchs=df['Price'].sum()
```

```python
PurchasingAnalysis=pd.DataFrame({'UniqueItemCount':[uniqueItems],
                                 'AveragePrice':[avgPrice],
                                 'NumberOfPurchases':[numPurchs],
                                 'SumOfPurchases':[sumPurchs]})
PurchasingAnalysis
```

```python
#other way
summary=(pd.DataFrame({"UniqueItemCount":len(df["Item ID"].unique()),
                       "USD_AveragePrice":[df["Price"].mean()],
                       "NumberOfPurchases":[df["Price"].count()],
                       "USD_SumOfPurchases": [df["Price"].sum()]            
                      })
        .round(2))
summary
```

```python
gcnt=Player_Count['Gender'].value_counts()
gpcnt=gcnt/TotalPlayers *100
GenderDemographics=pd.DataFrame({"TotalCount":gcnt,
                                "PercentPlayers":gpcnt})
GenderDemographics=GenderDemographics.round(2)
GenderDemographics
```

```python
#making the variable
gender_counts=(df.loc[:,["SN","Gender"]]
               .drop_duplicates()["Gender"]
               .value_counts())
```

```python
#with the variable
gender_demographix=(pd.DataFrame({"Total Count":gender_counts,
                                  "Percentage of Players":gender_counts/total_players*100})
                    .round(2))
gender_demographix
```

```python
#other way
gender_demografix=(pd.DataFrame({"Total Count":
                                  df.loc[:,["SN","Gender"]].drop_duplicates("SN")
                                  ["Gender"].value_counts(),
                                  "Percentage of Players":
                                  (df.loc[:,["SN","Gender"]].drop_duplicates("SN")
                                  ["Gender"].value_counts()
                                  /total_players*100)})
                    .round(2))
gender_demografix
```

```python
#other way
gender_data=(pd.DataFrame({"Purchase Count":df.groupby(["Gender"]).count()["Price"],
                          "Average Purchase Price (USD)":df.groupby(["Gender"]).mean()["Price"],
                           "Total Purchase Value (USD)":df.groupby(["Gender"]).sum()["Price"],
                           "Normalized Totals (USD)":
                           df.groupby(["Gender"]).sum()["Price"]/gender_demografix["Total Count"]})
            .round(2))
gender_data
```

```python
ageBins=[0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,999]
grpNames=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
```

```python
Player_Count["AgeRange"]=pd.cut(df["Age"],ageBins,labels=grpNames)
```

```python
ageDemoTotal=Player_Count['AgeRange'].value_counts()
ageDemoPecnt=ageDemoTotal/TotalPlayers *100
AgeDemographics=pd.DataFrame({"TotalCount":ageDemoTotal,
                                "PercentPlayers":ageDemoPecnt})
AgeDemographics=AgeDemographics.round(2)
AgeDemographics.sort_index()
```

```python
df["AgeRange"]=pd.cut(df["Age"],ageBins,labels=grpNames)
```

```python
agepurtot=df.groupby(["AgeRange"]).sum()["Price"]
ageavg=df.groupby(["AgeRange"]).mean()["Price"]
agecont=df.groupby(["AgeRange"]).count()["Price"]
ageavgpurchpp=agepurtot/AgeDemographics["TotalCount"]
```

```python
ageData=pd.DataFrame({"AgeTotSpent":agepurtot,
                      "AvgPurchPrice":ageavg,
                      "PurchCount":agecont,
                      "AvgSpentPerAgeGroup":ageavgpurchpp})
ageData=ageData.round(2)
ageData.sort_index()
```

```python
purchCont=df.groupby(["SN"]).count()["Price"]
avgPrce=df.groupby(["SN"]).mean()["Price"]
purchTtl=df.groupby(["SN"]).sum()["Price"]
```

```python
spndr=pd.DataFrame({"Purchase Count":purchCont,
                      "Average Purchase Price":avgPrce,
                      "Total Purchase Value":purchTtl})
spndr=spndr.sort_values("Total Purchase Value", ascending=False)
spndr=spndr.round(2)
spndr.head(5)
```

```python
itmDta=df.loc[:,["Item ID","Item Name","Price"]]
itmCont=itmDta.groupby(["Item ID","Item Name"]).count()["Price"]
itmPrc=itmDta.groupby(["Item ID","Item Name"]).mean()["Price"]
totPurchVal=itmDta.groupby(["Item ID","Item Name"]).sum()["Price"]
```

```python
mpopi=pd.DataFrame({"Purchase Count":itmCont,
                      "Item Price":itmPrc,
                      "Total Purchase Value":totPurchVal})
mpopi=mpopi.sort_values("Purchase Count", ascending=False)
mpopi=mpopi.round(2)
mpopi.head(5)
```
